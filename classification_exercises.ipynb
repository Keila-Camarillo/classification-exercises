Q: Is whether or not a customer churns independent of their choice on paperless billing?

$H_o$: churn is independent of paperless billing

$H_a$: churn is dependent of paperless billing

```python
observed = pd.crosstab(telco_train.paperless_billing_Yes, telco_train.churn)
```

```python
observed
```

```python
def chi2_test(table, variables=False):
    chi2, p, degf, expected = stats.chi2_contingency(table)
    print('Observed')
    print(table.values)
    print('\nExpected')
    print(table.astype(int))
    print('\n----')
    print(f'chi^2 = {chi2:.4f}')
    print(f'p     = {p}')
    if variables:
        return  chi2, p, degf, expected
    else:
        return
```

```python
chi2_test(observed)
```

```python
if p < alpha:
    print("We reject null hypothesis")
else:
    print("We fail to reject null hypothesis")
```

There is a some relationship between paperless billing and churned customers
